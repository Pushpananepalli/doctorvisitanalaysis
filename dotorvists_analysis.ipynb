# Multiplying 100000 to the 'income' column to update the income values

```python
import pandas as pd


df = pd.read_csv('Doctors.csv')

df['income'] = df['income'] *100000

df.to_csv('Doctors.csv', index=False)
df.head(10)
```

# Plotting a histogram of 'age'

```python
df = pd.read_csv('Doctors.csv')
plt.hist(df['age'], bins=20, color='green', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

# Plotting a Bar plot 

```python
# Grouping the DataFrame `df` by the 'gender' column and calculating the sum of the 'reduced' column for each grou
db= df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color = ['skyblue','pink'])
plt.title('Bar Chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```

# Plotting a scatter plot to explore the correlation between age and income

```python
df = pd.read_csv('Doctors.csv')
plt.figure(figsize=(4, 3))
plt.scatter(df['age'], df['income'], color='salmon', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Correlation between Age and Income')
plt.show()

```

# Creating a scatter plot to visualize the relationship between 'age' and 'illness' columns


```python
plt.figure(figsize=(4,4))
plt.scatter(x='age',y='illness',data=df)
plt.xlabel('age')
plt.ylabel('illeness')
```

# Plotting a box plot to compare the income distribution by gender

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='income', data=df, palette='pastel')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.title('Income Distribution by Gender')
plt.show()
```

# Creating a box plot to visualize the distribution of income values

# Pie charts to visualize the percentage of people getting different types of health insurance based on the 'freepoor', 'private', and 'freerepat' 

```python
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people getting government health insurance ")
plt.show()
 
Y = df[df['private']=='yes']
N = df[df['private']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people getting private health insurance ")
plt.show()
 
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people getting government health insurance due to oldage,disabillity or veteran status  ")
plt.show()

```
